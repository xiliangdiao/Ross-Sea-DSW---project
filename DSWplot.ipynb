```python
data =  xr.open_dataset('C:\\Users\\DELL\\Documents\\xiliangdiao\\vol_trans_across_isobath_binned_2010year.nc')
```

```python
###### loading and computations for isobath transport ##########
## load control + take mean in time (control year 5 and 10 tested and pretty much identical)
vol_trans_across_isobath_binned_2010 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_2010year.nc')
vol_trans_across_isobath_binned_2010 = vol_trans_across_isobath_binned_2010.sum('contour_index')
vol_trans_across_isobath_binned_2010 = vol_trans_across_isobath_binned_2010.vol_trans_across_isobath_binned
## load experiment + take mean in time
vol_trans_across_isobath_binned_2020 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_2020year.nc')
vol_trans_across_isobath_binned_2020 = vol_trans_across_isobath_binned_2020.sum('contour_index')
vol_trans_across_isobath_binned_2020 = vol_trans_across_isobath_binned_2020.vol_trans_across_isobath_binned
## load experiment + take mean in time
vol_trans_across_isobath_binned_2030 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_2030year.nc')
vol_trans_across_isobath_binned_2030 = vol_trans_across_isobath_binned_2030.sum('contour_index')
vol_trans_across_isobath_binned_2030 = vol_trans_across_isobath_binned_2030.vol_trans_across_isobath_binned
## load experiment + take mean in time
vol_trans_across_isobath_binned_2040 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_2040year.nc')
vol_trans_across_isobath_binned_2040 = vol_trans_across_isobath_binned_2040.sum('contour_index')
vol_trans_across_isobath_binned_2040 = vol_trans_across_isobath_binned_2040.vol_trans_across_isobath_binned
## load experiment + take mean in time
vol_trans_across_isobath_binned_1850 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_1850.nc')
vol_trans_across_isobath_binned_1850 = vol_trans_across_isobath_binned_1850.sum('contour_index')
vol_trans_across_isobath_binned_1850 = vol_trans_across_isobath_binned_1850.vol_trans_across_isobath_binned
## load control + take mean in time (control year 5 and 10 tested and pretty much identical)
vol_trans_across_isobath_binned_2050 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_2050year.nc')
vol_trans_across_isobath_binned_2050 = vol_trans_across_isobath_binned_2050.sum('contour_index')
vol_trans_across_isobath_binned_2050 = vol_trans_across_isobath_binned_2050.vol_trans_across_isobath_binned
## load experiment + take mean in time
vol_trans_across_isobath_binned_2060 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_2060year.nc')
vol_trans_across_isobath_binned_2060 = vol_trans_across_isobath_binned_2060.sum('contour_index')
vol_trans_across_isobath_binned_2060 = vol_trans_across_isobath_binned_2060.vol_trans_across_isobath_binned
## load experiment + take mean in time
vol_trans_across_isobath_binned_2070 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_2070year.nc')
vol_trans_across_isobath_binned_2070 = vol_trans_across_isobath_binned_2070.sum('contour_index')
vol_trans_across_isobath_binned_2070 = vol_trans_across_isobath_binned_2070.vol_trans_across_isobath_binned
## load experiment + take mean in time
vol_trans_across_isobath_binned_2080 =  xr.open_dataset('C:\\Users\\DELL\\Documents\\newplots\\vol_trans_across_isobath_binned_2080year.nc')
vol_trans_across_isobath_binned_2080 = vol_trans_across_isobath_binned_2080.sum('contour_index')
vol_trans_across_isobath_binned_2080 = vol_trans_across_isobath_binned_2080.vol_trans_across_isobath_binned

## sum along the isobath

# ## cumsum from the bottom

net_offshore_trans_control = np.cumsum(vol_trans_across_isobath_binned_2010[::-1])[::-1]* 1e-6

net_offshore20 = np.cumsum(vol_trans_across_isobath_binned_2020[::-1])[::-1]* 1e-6

net_offshore30 = np.cumsum(vol_trans_across_isobath_binned_2030[::-1])[::-1]* 1e-6

net_offshore40 = np.cumsum(vol_trans_across_isobath_binned_2040[::-1])[::-1]* 1e-6

net_offshore50 = np.cumsum(vol_trans_across_isobath_binned_2050[::-1])[::-1]* 1e-6
####################################################################################################
net_offshore60= np.cumsum(vol_trans_across_isobath_binned_2060[::-1])[::-1]* 1e-6

net_offshore70= np.cumsum(vol_trans_across_isobath_binned_2070[::-1])[::-1]* 1e-6

net_offshore80= np.cumsum(vol_trans_across_isobath_binned_2080[::-1])[::-1]* 1e-6

net_offshore18= np.cumsum(vol_trans_across_isobath_binned_1850[::-1])[::-1]* 1e-6

```

```python
def DSW_transport_full(ax, control = net_offshore_trans_control):
    ax.plot(net_offshore18,data.isopycnal_bins,color = 'k', linestyle = '-.', linewidth = 2, label = '1850')
    ax.plot(net_offshore_trans_control,data.isopycnal_bins,color = 'k', linewidth = 2, label = '2010')
    ax.plot(net_offshore20,data.isopycnal_bins, color = 'darkred', linewidth = 2, label = '2020')
    ax.plot(net_offshore30,data.isopycnal_bins,color = 'purple', linewidth = 2, label = '2030')
    ax.plot(net_offshore40,data.isopycnal_bins,color = 'red', linestyle = '--', linewidth = 2, label = '2040')
    ax.plot(net_offshore50,data.isopycnal_bins,color ='darkgreen',linestyle = '--',linewidth = 2, label = '2050')
    
    ax.plot(net_offshore60,data.isopycnal_bins,color = 'darkgreen', linestyle = ':', linewidth = 2, label = '2060')
    ax.plot(net_offshore70,data.isopycnal_bins, color = 'blue', linestyle = ':', linewidth = 2, label = '2070')
    ax.plot(net_offshore80,data.isopycnal_bins,color = 'blue', linestyle = '--', label = '2080')
    #ax.plot(net_offshore18,data.isopycnal_bins,color = 'k', linestyle = '--', linewidth = 2, label = '1850')
    #ax.plot(net_offshore60,data.isopycnal_bins,color ='darkred',linewidth = 2, label = 'M (Year 10)')
    
    ax.set_ylabel(r'$\sigma_{1}$ (kg m$^{-3}$)', fontsize = 35)
    ax.set_xlabel('Offshore transport across 1000m isobath (Sv)', fontsize = 35)
    ax.plot([0,0],[30,32.8],'k',linewidth=0.5)
    ax.set_xticks([0,5,10,15,20])
    ax.set_xticklabels(['0','5','10','15','20'], fontsize = 28)
    ax.set_yticks([31.6,31.7,31.8,31.9,32.0,32.1,32.2,32.3,32.4,32.5,32.6,32.7,32.8])
    ax.set_yticklabels(['31.6','31.7','31.8','31.9','32.0','32.1','32.2','32.3','32.4','32.5','32.6','32.7','32.8'], fontsize = 28)
    #ax.set_xlim((-1,11))
    ax.set_ylim((32.8,31.6))
    ax.plot([-3,13],[32.53, 32.53],'k',linewidth=3)
    ax.plot([-3,13],[32.37, 32.37],'k',linewidth=3)
    ax.legend(fontsize = 16) 
```
